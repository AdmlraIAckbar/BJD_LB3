```python
V = cur_data[1] * cur_data[2] * cur_data[3] # считаем объем
B1000 = 0  # инициализируем B1000 в зависимости от варианта и типа помещения
if variant == 0: B1000 = V/20
elif variant == 1: B1000 = V/20
elif variant == 2: B1000 = V/20
elif variant == 3: B1000 = V/6
elif variant == 4: B1000 = V/6
elif variant == 5: B1000 = V/10
elif variant == 6: B1000 = V/6
elif variant == 7: B1000 = V/6
elif variant == 8: B1000 = V/10
elif variant == 9: B1000 = V/6

# инициализируем u u зависимости от V
if V<200: u = table1[0]
elif V>1000: u = table1[1]
else: u = table1[2]

B = []# инициализируем u B зависимости от V
for i in range(len(u)-1):
    B.append(u[i+1] *B1000)
# вычисляем Lp 
for i in range(cur_data[4]):
    r = cur_data[6+i]
    #temp = r/lmx
    temp = 1
    X=1# на кафедре сказали что 1 а я и не против
Lp = []
for j in range(8):
    temp_lp1 = 0
    temp_lp2 = 0
    for n in range(cur_data[4]):
        strr = cur_data[5]
        #temp = r/lmx
        #temp = 2 # просто от балды спросить как правильно
        #if temp<0.5: X=4 #temp=0
        #elif temp<1: X=3.5#temp=0.7
        #elif temp<1.5: X=2#temp=1
        #else: X=1#temp = 1.5
        index = -1
        temp = strr.split(' ')[n].split('-')[1]
        for i in range(len(table3)):
            S = 4 * 3.14 * cur_data[6+n]
            if table3[n][0] == temp: index = n
            temp_lp1+= ( (10**(0.1*table3[index][j+1]))* X* 1)/(S)
            temp_lp2+= 10**(0.1*table3[index][j+1])
    temp_lp = 10*math.log10(temp_lp1 + 4/B[j]*temp_lp2)
    Lp.append(round(temp_lp))
    
#Lp
```
